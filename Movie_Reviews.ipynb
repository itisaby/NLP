```python
from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
    
```

```python
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
import string
punc = list(string.punctuation)
stops.update(punc)
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = Lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(document), category) for document, category in documents]
```

```python
documents[0]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words+=doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
```

```python
common = freq.most_common(3000)
features = [i[0] for i in common]
features
```

```python
def get_features_dict(words):
    current_features ={}
    word_set = set(words)
    for w in features:
        current_features[w] = w in word_set 
    return current_features
```

```python
output = get_features_dict(training_documents[0][0])
output
```

```python
training_data = [(get_features_dict(doc), category) for doc, category in training_documents]
```

```python
testing_data = [(get_features_dict(doc), category) for doc, category in testing_documents]
```

```python
from nltk import NaiveBayesClassifier 
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```

```python
classifier.show_most_informative_features(15)
```
